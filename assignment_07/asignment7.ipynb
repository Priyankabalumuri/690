# Use all countries, and pick year 2020, make a scatter plot with x being GDP per Capita and y being Life Expentancy. This plot will tell you the relationship between wealth and health.

```python
df5 = df.query(' Year == 2020')
df5
```

```python
hw = ["SP.DYN.LE00.IN", "NY.GDP.PCAP.PP.CD"]
df6 = df5[df5["indicator"].isin(hw)]
df6.shape
```

```python
final_df = pd.pivot_table(df6, values = 'value', index=['Country Code', 'Country Name','Year'], columns = 'indicator').reset_index()
final_df.head()
```

```python
fig = px.scatter(final_df,x="NY.GDP.PCAP.PP.CD",y="SP.DYN.LE00.IN", title="Wealth vs Health", color="Country Name",)
fig.show()
```
