6.Cleanse the data to make the year, population, cost, cost per person numerical type

**cleanse_year**

```python
def cleanse_year(a_string):
    return int(a_string.replace("*",""))    # removing '*' present in the column and converting to int
```

```python
df["Census Year"] = df["Census Year"].apply(cleanse_year)
df.head(10)
```

**cleanse_population**

```python
def cleanse_population(a_string):
    return int(a_string.replace(",",""))   # removing ',' in population column
```

```python
df["Total Population"] = df["Total Population"].apply(cleanse_population)
df.head(10)
```

**cleanse_cost**

```python
def cleanse_cost(a_string):
  if '$' in a_string and 'Billion' in  a_string:
    a_string = float(a_string.replace("$","").replace("Billion",""))
    return int(a_string * 1000000000) 
  return int(a_string.replace("$","").replace(",",""))
```

```python
df["Census Cost"] = df["Census Cost"].apply(cleanse_cost)
df.head(10)
```

**cleanse_cost_per_person**

```python
def cleanse_cost_per_person(a_string):
  if "cents" in a_string:
    a_string=a_string.replace("cents","")
    a_string=(float(a_string))/100
    return a_string
  else:
    return float(a_string.replace("$",""))
```

```python
df["Average Cost Per Person"] = df["Average Cost Per Person"].apply(cleanse_cost_per_person)
df.head(10)
```
